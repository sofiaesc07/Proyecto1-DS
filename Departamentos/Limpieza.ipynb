```python
#librerias a importar
import pandas as pd
import numpy as np
```

```python
#Se importan los csv de cada uno de los departamentos
ALTA_VERAPAZ = pd.read_csv('ALTA_VERAPAZ.csv')
BAJA_VERAPAZ = pd.read_csv('BAJA_VERAPAZ.csv', encoding='latin-1')
CIUDAD_CAPITAL = pd.read_csv('CIUDAD_CAPITAL.csv')
GUATEMALA = pd.read_csv('GUATEMALA.csv', encoding='latin-1')
CHIMALTENANGO = pd.read_csv('CHIMALTENANGO.csv', encoding='latin-1')
CHIQUIMULA = pd.read_csv('CHIQUIMULA.csv', encoding='latin-1')
EL_PROGRESO = pd.read_csv('EL PROGRESO.csv', encoding='latin-1')
ESCUINTLA = pd.read_csv('ESCUINTLA.csv', encoding='latin-1')
HUEHUETENANGO = pd.read_csv('HUEHUETENANGO.csv', encoding='latin-1')
IZABAL = pd.read_csv('IZABAL.csv', encoding='latin-1')
JALAPA = pd.read_csv('JALAPA.csv', encoding='latin-1')
JUTIAPA = pd.read_csv('JUTIAPA.csv', encoding='latin-1')
```

```python
#Se concatenan todos los dataframes --- agregar los demas departamentos
df = pd.concat([ALTA_VERAPAZ, CIUDAD_CAPITAL, GUATEMALA, CHIMALTENANGO,CHIQUIMULA, EL_PROGRESO, ESCUINTLA, 
                BAJA_VERAPAZ, HUEHUETENANGO, IZABAL, JALAPA, JUTIAPA], axis=0, ignore_index=True)
df
```

```python
#Listamos los nombres de las columnas
list(df.keys())

#Borramos filas duplicadas
df=df.drop_duplicates()
```

```python
#Observamos la cantidad de na de cada variable
print('CODIGO: ', df['CODIGO'].isna().sum())
print('DISTRITO: ', df['DISTRITO'].isna().sum())
print('DEPARTAMENTO: ', df['DEPARTAMENTO'].isna().sum())
print('MUNICIPIO: ', df['MUNICIPIO'].isna().sum())
print('ESTABLECIMIENTO: ', df['ESTABLECIMIENTO'].isna().sum())
print('DIRECCION: ', df['DIRECCION'].isna().sum())
print('TELEFONO: ', df['TELEFONO'].isna().sum())
print('SUPERVISOR: ', df['SUPERVISOR'].isna().sum())
print('DIRECTOR: ', df['DIRECTOR'].isna().sum())
print('NIVEL: ', df['NIVEL'].isna().sum())
print('SECTOR: ', df['SECTOR'].isna().sum())
print('AREA: ', df['AREA'].isna().sum())
print('STATUS: ', df['STATUS'].isna().sum())
print('MODALIDAD: ', df['MODALIDAD'].isna().sum())
print('JORNADA: ', df['JORNADA'].isna().sum())
print('PLAN: ', df['PLAN'].isna().sum())
print('DEPARTAMENTAL: ', df['DEPARTAMENTAL'].isna().sum())
```

```python
#Botamos todas las filas con valores nulos en el código o establecimiento ya que estas deberían de ser unicas
df=df.dropna(subset = ['CODIGO', 'ESTABLECIMIENTO'])
df
```

```python
#Contamos las repeticiones de establecimientos
df.ESTABLECIMIENTO.value_counts()
```

```python
#Removemos acentos
cols = df.select_dtypes(include=[object]).columns
df[cols] = df[cols].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
df.ESTABLECIMIENTO.value_counts()
```

#### Variable Departamento

Ciudad Capital tenian sus datos distribuidos de una manera distinta pues la variables DEPARTAMENTO guarda el municipio del establecimiento y la variable MUNICIPIO guarda la zona de la capital en la que recide, así que se crea una variable para reordenar estos datos y volver a la ciudad capital parte de guatemala

```python
#creamos la variable zona para la ciudad capital
df = df.assign(ZONA = lambda x: np.where(x.DEPARTAMENTO == "CIUDAD CAPITAL", x.MUNICIPIO, None))
#volvemos el departamento un municipio 
df = df.assign(MUNICIPIO = lambda x: np.where(x.DEPARTAMENTO == "CIUDAD CAPITAL", x.DEPARTAMENTO, x.MUNICIPIO))
#unimos ciudad capital a guatemala
df = df.assign(DEPARTAMENTO = lambda x: np.where(x.DEPARTAMENTO == "CIUDAD CAPITAL", "GUATEMALA", x.DEPARTAMENTO))
```

#### Variable dirección

Esta variable tiene diferentes tipos de formatos, decidimos validar direcciones claramente erroneas, constituidas por un par decaracteres invalidos, como "-", "--", "---", ".", "". y las definimos como NONE. 

```python
setFaltasDireccion = ['-', '--', '---', '.', ' ', '  ']
df.DIRECCION = df.DIRECCION.replace(setFaltasDireccion, None)
```

#### Variable Telefono

A muchos numeros telefonicos les faltan numeros, fueron guardados con decimales o inculo algunos establecimientos cuentan con más de un numero telefonico

```python
#i una entrada posee 8 caracteres o más, botamos la cola del dato y nos quedamos con los primeros 8 caracteres. 
# Si es menor a 8 o posee caracteres es desechado

df['TELEFONO'] = df['TELEFONO'].apply(lambda x: str(x))
df['TELEFONO'] = df['TELEFONO'].apply(lambda x: x[0:8] if len(x) >= 8  else None)
```

#### Variable Director

```python
#Contamos las repeticiones de directores
print(df.DIRECTOR.value_counts())

#Convertimos en NaN todos 
df['DIRECTOR']=df['DIRECTOR'].replace('.',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('..',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('...',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('-',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('--',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('---',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('----',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('-----',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('------',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('-------',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('--------',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('---------',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('----------',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('-----------',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('------------',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('-------------',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('--------------',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('---------------',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('----------------',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('-----------------',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('-------------------',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('0',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('X',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace(' ',np.NaN)
df['DIRECTOR']=df['DIRECTOR'].replace('SIN DATO',np.NaN)

df.DIRECTOR.value_counts()
```
